## TRAINING AND TESTING THE MODEL

```python
x = df2.drop(["Species","Id"],axis=1)
y= df2["Species"]
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test =train_test_split(x,y,test_size=0.25,random_state=1)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

## ALGORITHM

```python
from sklearn.tree import DecisionTreeClassifier,plot_tree

df2 = DecisionTreeClassifier(max_depth=5)
df2.fit(x_train, y_train)
```

#### 1.we have taken a depth of the tree as (max_depth=5). if the depth of the tree increases it may result into the overfitting problem .
#### 2.sklearn uses the gini impurity which is quite similar to the entropy but the advantage of gini impurity is its has more computational efficient.


 ## DECISION TREE MODEL

```python
df=df2.fit(x,y)
plt.figure(figsize=(20,30))
plot_tree(df,filled=True)
```

## ACCURACY

```python
df2.score(x_test,y_test)
```

## PREDICTED VALUES

```python
predicted_value=df2.predict(x_test)
df3=pd.DataFrame({"ACTUAL VALUES": y_test,"PREDICTED VALUES":predicted_value})
df3.head(10)
```
